```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np

head = []
head_inv = []
q = []
with open('epsilon.out') as f:
    for l in f:
        if 'Head of Epsilon' in l:
            if 'Inverse' in l:
                head_inv.append(float(l.split()[7]))
            else:
                head.append(float(l.split()[6]))
        if '|Q0|' in l or '|Q|' in l:
            q.append(float(l.split()[-1]))
    f.close()
    
fig = plt.figure(figsize=(8,6))
matplotlib.rc('font', size=14.0)
plt.plot(q,head,'bo',label='$\epsilon_{00}(q)$')
plt.plot(q,1/np.array(head_inv),'ro',label='$1/\epsilon_{00}^{-1}(q)$')
plt.legend()
plt.xlabel('|q|')
plt.ylabel('Dielectric function')

#Uncomment to save the figure
# plt.savefig('Head_Epsilon.png')
```
